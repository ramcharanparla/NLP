```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from pprint import pprint
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
sentences=['Nothing much to say as it is a macbook. the M1 processor works like a charm',
           'Amazing latop, super performance with M1, its blazzing fast',
           'Working very slow and takes 15-20 minutes to start thus not worth for money',
           'This is not a good laptop. It is vey slow. It is taking 20 minutes to start']
```

```python
sentence_embeddings = model.encode(sentences)
```

```python
for sentence, embedding in zip(sentences, sentence_embeddings):
  print("Sentence:",sentence)
  print('Embeddings:', embedding)
```

```python
pprint(f'similarity between {sentences[0]} and {sentences[1]} is {cosine_similarity(sentence_embeddings[0].reshape(1,-1), sentence_embeddings[1].reshape(1,-1))[0][0]}')
```

```python
pprint('Similarity between {} and {} is {}'.format(sentences[0],
                                                    sentences[2],
                                                    cosine_similarity(sentence_embeddings[0].reshape(1, -1),
                                                                      sentence_embeddings[2].reshape(1,-1))[0][0]))
```
