```python
import pandas as pd
import librosa
import soundfile as sf
import tensorflow as tf
from sklearn.model_selection import train_test_split


# Creating directories for audio and transcriptions
os.makedirs(f"{save_path}/audio", exist_ok=True)

# Storing audio paths and transcriptions
data = []
for idx, sample in enumerate(dataset.select(range(500))):
    data.append({"audio_path": sample["audio"]["path"], "transcript": sample["text"]})

# Converting to DataFrame
df = pd.DataFrame(data)

# Save transcriptions
transcript_path = f"{save_path}/transcripts.csv"
df.to_csv(transcript_path, index=False)
print("Extracted audio paths and transcriptions saved.")

# Function for converting audio to 16kHz mono WAV format
def convert_audio(input_path, output_path):
    y, sr = librosa.load(input_path, sr=16000, mono=True)  # Convert to 16kHz mono
    sf.write(output_path, y, 16000)

# Convert all audio files
for idx, row in df.iterrows():
    output_path = f"{save_path}/audio/{idx}.wav"
    convert_audio(row["audio_path"], output_path)
print("Audio files converted to 16kHz WAV format.")

# Split dataset into Train, Validation, Test
train, temp = train_test_split(df, test_size=0.2, random_state=42)
val, test = train_test_split(temp, test_size=0.5, random_state=42)

# Save split datasets
train.to_csv(f"{save_path}/train.csv", index=False)
val.to_csv(f"{save_path}/val.csv", index=False)
test.to_csv(f"{save_path}/test.csv", index=False)

print("All CSV files saved in the ami_dataset folder.")
```

```python
import os
import re
import librosa
import numpy as np
import pandas as pd
import tensorflow as tf
from datasets import load_dataset, Dataset
from transformers import Wav2Vec2Processor, TFAutoModel, AutoProcessor
from transformers import DataCollatorWithPadding, create_optimizer
from jiwer import wer
from sklearn.model_selection import train_test_split
```

```python
# Load TensorFlow-Compatible Model
model_name = "openai/whisper-small"

# Load processor
processor = AutoProcessor.from_pretrained(model_name)

# Load the model (TensorFlow version)
model = TFAutoModel.from_pretrained(model_name)

print("TensorFlow Model and Processor Loaded Successfully!")
```

```python
# Function to preprocess audio
def preprocess_audio(file_path):
    """Load audio file and extract padded input features for Whisper."""
    speech, sample_rate = librosa.load(file_path, sr=16000, mono=True)

    # Extract input features (spectrogram)
    input_features = processor(
        speech, sampling_rate=16000, return_tensors="np"
    ).input_features

    return input_features.squeeze(0).tolist()  # Convert to list for Dataset storage

# Function to preprocess text
def preprocess_text(text):
    """Clean and tokenize transcript text with padding."""
    text = text.lower()
    text = re.sub(r"[^\w\s']", "", text)  # Remove punctuation except apostrophes

    # Tokenize text with padding
    labels = processor.tokenizer(
        text, return_tensors="np", padding="max_length", truncation=True, max_length=128
    ).input_ids

    labels = labels.squeeze(0).tolist()

    # Convert padding token to -100 for loss computation
    labels = [-100 if token == processor.tokenizer.pad_token_id else token for token in labels]

    return labels
```

```python
# Load dataset from CSV
df = pd.read_csv("/content/ami_dataset/train.csv")

# Apply preprocessing
df["input_features"] = df["audio_path"].apply(preprocess_audio)
df["decoder_input_ids"] = df["transcript"].apply(preprocess_text)

# Remove unused columns
# df = df.drop(columns=["audio_path", "transcript"])

# Convert DataFrame into Hugging Face Dataset
dataset = Dataset.from_pandas(df)

print("Dataset successfully preprocessed and converted!")
df.columns
```

```python
def tf_generator():
    for sample in dataset:
        input_features = np.array(sample["input_features"], dtype=np.float32)

        if input_features.shape[0] != 80:
            input_features = input_features.T

        yield {
            "input_features": input_features,
            "decoder_input_ids": np.array(sample["decoder_input_ids"], dtype=np.int32),
            "decoder_attention_mask": np.ones_like(sample["decoder_input_ids"], dtype=np.int32),
        }

tf_dataset = tf.data.Dataset.from_generator(
    tf_generator,
    output_signature={
        "input_features": tf.TensorSpec(shape=(80, None), dtype=tf.float32),
        "decoder_input_ids": tf.TensorSpec(shape=(None,), dtype=tf.int32),
        "decoder_attention_mask": tf.TensorSpec(shape=(None,), dtype=tf.int32),
    }
).padded_batch(2, padding_values={"input_features": 0.0, "decoder_input_ids": -100, "decoder_attention_mask": 0}).prefetch(tf.data.AUTOTUNE)

```

```python
import tensorflow as tf
from tensorflow.keras import mixed_precision

mixed_precision.set_global_policy('mixed_float16')  # Enables lower memory usage
```

```python
import tensorflow as tf
tf.keras.backend.clear_session()  # Clears memory leaks before training
```

```python
for batch in tf_dataset.take(1):
    for key, value in batch.items():
        print(f"{key}: {value.shape}")
```

```python
# # Loss Function and Optimizer
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True, reduction="none")
optimizer = tf.keras.optimizers.Adam(learning_rate=3e-4)


# # Compile Model
model.compile(optimizer=optimizer, loss=loss_fn, metrics=["accuracy"])

# # Train the model
model.fit(
    tf_dataset.map(lambda x: (x, x["decoder_input_ids"])),
    epochs=3
)

# # Save Model
model.save_pretrained("whisper-tf")
processor.save_pretrained("whisper-tf")

print("Training completed and model saved!")
```
